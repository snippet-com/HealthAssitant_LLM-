```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris= sns.load_dataset('iris')
```

```python
print(iris)
```

**SCATTERPLOT**

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width')
plt.show()
```

```python
sns.scatterplot(data= iris, x='petal_length', y='petal_width', hue='species')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data= iris,
    x= 'sepal_length',
    y='sepal_width',
    hue='species',
    style='species',
    s=100
)

plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

```





**HISTOGRAM**






```python
iris['sepal_length'].hist()
plt.show()
```

```python
sns.histplot(iris['sepal_length'], kde=True)
plt.show()
```

```python
sns.histplot(data=iris, x='petal_length', hue='species', kde=True)
plt.show()
```

BOX **PLOT**

```python
sns.boxplot(x=iris['sepal_length'])
plt.show()
```

```python
sns.boxplot(data=iris, x='species', y='petal_length')
plt.show()
```


**Summary:**

*   Petal features are the most important variables for distinguishing between species
*   Sepal features are less reliable for classification due to overlap

*   The dataset is clean and well-structured, making it suitable for machine learning tasks
*   One class (setosa) is easily separable, while others require more careful analysis
